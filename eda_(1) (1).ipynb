```python
df = df.astype({'BudgetInNum':'float'})
print(df.dtypes)
```

```python
x = round(df["BudgetInNum"].mean())
df["BudgetInNum"].fillna(x, inplace = True)
df.info()
```

```python
ind=df['Budget'].isnull()
indx = df[ind].index
indx
```

```python
for i in df['Budget'].index:
    if i in indx:
            df['Budget'].fillna(value=('$'+str.replace(str(df['BudgetInNum'][i]),'.0','')+' crore'), inplace= True)
df
```

```python
for i in df['Budget'].index:
    if i in indx:
            df['Budget'].fillna(value=('$'+str.replace(str(df['BudgetInNum'][i]),'.0','')+' crore'), inplace= True)
df
```

```python
df['Gross in India (INR crore)'] = df['Gross in India (INR crore)'].str.replace(',','')
```

```python
x = 0
c = 0
for i in df['Gross in India (INR crore)']:
    if str(i) != 'nan':
        x = x + float(i)
    c = c+1
avg = round(x/c,2)
avg
```

```python
df['Gross in India (INR crore)'].fillna(value = avg,inplace = True)
```

```python
std=np.std(df['BudgetInNum'],ddof=1)
mean=np.mean(df['BudgetInNum'])
mean
```

```python
domain=np.linspace(np.min(df['BudgetInNum']),np.max(df['BudgetInNum']))
plt.plot(domain,norm.pdf(domain,mean,std))
plt.hist(df['BudgetInNum'],edgecolor='black',alpha=.5,density=True)
plt.title("Normal Fit")
plt.xlabel("value")
plt.ylabel("density")
plt.show()
```
